```python
# Function for plotting data
def plot_data(data, name, bin_size=2, color=None, r=(0,120), limit=28000):
    bins = int(120/bin_size)
    plt.figure(figsize=(15,10))
    plt.hist(data, bins=bins, color=color, range=(0,120))
    plt.title('Distribution of ' + name + ' speed')
    plt.ylabel('Frequency')
    plt.ylim((0,limit))
    plt.xlabel('speed(km/h)')
    plt.axvline(np.mean(data), color='black', linestyle='dashed', linewidth=1)
    plt.axvline(np.median(data), color='red', linestyle='dashed', linewidth=1)
    plt.axvline(np.percentile(data, 25), color='red', linestyle='dashed', linewidth=1)
    plt.axvline(np.percentile(data, 75), color='red', linestyle='dashed', linewidth=1)
    plt.axvline(np.max(data), color='green', linestyle='dashed', linewidth=1)
    print('Mean(black): {:>4.4f}, Q1(red): {:>4.4f}, Median(red): {:>4.4f}, Q3(red): {:>4.4f}, Maximum(green): {:>4.4f}'.format(np.mean(data), np.percentile(data, 25), np.median(data), np.percentile(data, 75), np.max(data)))
    
```

```python
# Plot Car data
plot_data(data['car'], 'car', color='#2096F3')
```

```python
# Plot data for bike

plot_data(data['bike'], 'bike', color='#4CAF50')
```

```python
# Function to plot comparison

def plot_comparison(data1, data2, name1, name2, color1=None, color2=None, bin_size=2, r=(0,120), limit=28000):
    bins = int(120/bin_size)
    plt.figure(figsize=(15,10))
    n, bins, patches = plt.hist(data1, bins=bins, range=(0,120), alpha=1, color=color1)
    n, bins, patches = plt.hist(data2, bins=bins, alpha=0.5, color=color2)
    plt.title(name1 + ' - ' + name2 + ' Comparison (' + name1 + ': solid ' + name2 + ': transparent)')
    plt.ylabel('Frequency')
    plt.ylim((0,limit))
    plt.xlabel('Speed(km/h)')
```

```python
plot_comparison(data['tram'], data['bike'], 'Tram', 'Bike', color1='#FF5723', color2='#4CAF50')
```
